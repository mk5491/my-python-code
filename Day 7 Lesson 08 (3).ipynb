```python
'''
Format
This data frame contains the following columns:

crim
per capita crime rate by town.

zn
proportion of residential land zoned for lots over 25,000 sq.ft.

indus
proportion of non-retail business acres per town.

chas
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

nox
nitrogen oxides concentration (parts per 10 million).

rm
average number of rooms per dwelling.

age
proportion of owner-occupied units built prior to 1940.

dis
weighted mean of distances to five Boston employment centres.

rad
index of accessibility to radial highways.

tax
full-value property-tax rate per \$10,000.

ptratio
pupil-teacher ratio by town.

black
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

lstat
lower status of the population (percent).

medv
median value of owner-occupied homes in \$1000s.
'''
```

```python
# step 2 - splitting x and y variables
import pandas as pd
x=pd.DataFrame(Boston['rm'])
y=Boston['medv']
```

```python
# Step 3 - Running model
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
Model_Boston = lm.fit(x,y) 
#or
Model_Boston = LinearRegression().fit(x,y) 
```

```python
print(Model_Boston.intercept_)
print(Model_Boston.coef_)
```

```python
# Step 4 -  Making predictions 
pred_medv = Model_Boston.predict(x)
len(pred_medv)
```

```python
Boston_predicted = pd.concat([Boston,pd.DataFrame(pred_medv,columns=['Predicted_price'])],axis=1)
Boston_predicted
```

```python
Boston_predicted['Error_pct'] = abs(Boston_predicted.medv-Boston_predicted.Predicted_price)/Boston_predicted.medv
```

```python
Boston_predicted
```

```python
Boston_predicted['Error_pct'].mean()
```

```python
# accuracy rate
1- Boston_predicted['Error_pct'].mean()
```

```python
######################## Simple linear regression with Train and Test data split ########################
# step 2 - splitting x and y variables
import pandas as pd
x=pd.DataFrame(Boston['rm'])
y=Boston['medv']
```

```python
# Step 2-  Important. Sampling. NOTE – function “train_test_split” 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size=.33)
```

```python
print(x_test.shape)
print(y_test.shape)
print(x_train.shape)
print(y_train.shape)
```

```python
# Step 3 - Running model in Training data
from sklearn.linear_model import LinearRegression
Model_Boston = LinearRegression().fit(x_train,y_train) 
```

```python
print(Model_Boston.intercept_)
print(Model_Boston.coef_)
```

```python
# Step 4 -  Making predictions 
pred_medv = Model_Boston.predict(x_test)
len(pred_medv)
```

```python
Final_test_data = pd.concat([pd.concat([x_test,y_test],axis=1).reset_index(),pd.DataFrame(pred_medv,columns=['predicted'])],axis=1)
Final_test_data
```

```python
Final_test_data['Error_pct'] = abs(Final_test_data.medv-Final_test_data.predicted)/Final_test_data.medv
```

```python
Final_test_data.Error_pct.mean()
```

```python
1-Final_test_data.Error_pct.mean()
```

```python
######################## multiple linear regression –  ########################
import pandas as pd
x=Boston.drop('medv',axis=1)
y=Boston['medv']

from sklearn.linear_model import LinearRegression 
lm = LinearRegression()
boson_equation = lm.fit(x,y) 
# equation beta coefficients
print(boson_equation.intercept_)
print(x.columns)
print(boson_equation.coef_)
```

```python
boson_equation.score(x,y)
```

```python
## making predictions
Predicted_values = boson_equation.predict(x)
Predicted_values
Predicted_values = pd.DataFrame(Predicted_values,columns=['Predicted Price of house'])
```

```python
## adding predicted values to original data
Predicted_Price_table = pd.concat([x,y,Predicted_values],axis=1)
Predicted_Price_table

### finding out error percentages
Predicted_Price_table['Residual'] = abs(Predicted_Price_table.medv - Predicted_Price_table['Predicted Price of house'])/Predicted_Price_table.medv
Predicted_Price_table
1-Predicted_Price_table.Residual.mean()
```
