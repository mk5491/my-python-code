```python
# activity - Create frequency of math and science scores and put that in stacked bar plot
```

```python
df_math = pd.DataFrame(hsb.groupby(['math']).agg({'math':'count'}))
df_math
df_science = pd.DataFrame(hsb.groupby(['science']).agg({'science':'count'}))
df_science
Final = pd.concat([df_math,df_science],axis=1)
Final.math = Final.math.fillna(0)
Final.science = Final.science.fillna(0)
Final

plt.figure(figsize=(16,4))
plt.bar(Final.index,Final.math,color='r')
plt.bar(Final.index,Final.science,bottom=Final.math)
```
