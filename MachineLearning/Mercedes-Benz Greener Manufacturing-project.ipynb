```python
# assign X abd Y values from dataser
x_train = df_train_filter.drop(['ID', 'y'], axis=1)
y_train = df_train_filter['y']

x_test = df_test_filter.drop(['ID'], axis=1)
```

```python
x_train.shape, y_train.shape, x_test.shape
```

```python
x_train.info()
```

```python
# apply label encoder for object datatype columns 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for i in x_train.dtypes[x_train.dtypes == "object"].index:
    x_train[i]=le.fit_transform(x_train[i])
```

```python
x_train.info()
```

```python
for i in x_test.dtypes[x_test.dtypes == "object"].index:
    x_test[i]=le.fit_transform(x_test[i])
```

```python
x_test.info()
```

```python
# apply standard scaler for normalising the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
scaler.fit(x_test)
```

```python
scaled_data_train = scaler.transform(x_train)
scaled_data_test = scaler.transform(x_test)
```

```python
# apply dimenson reduction with PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=90)
```

```python
pca.fit(scaled_data_train)
pca.fit(scaled_data_test)
```

```python
x_train_pca = pca.transform(scaled_data_train)
x_test_pca = pca.transform(scaled_data_test)
```
